```python
#Finding no.of balls faced by each batsman
balls = deliv_df.groupby('batsman')['ball'].count().reset_index()
balls.columns=['batsman', 'balls']
balls
```

```python
runs = deliv_df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
runs
```

```python
four = deliv_df[deliv_df['batsman_runs']==4]
```

```python
#Number of fours hit by each batsman
runs_4 = four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_4.columns = ['batsman', 'fours']
runs_4
```

```python
six = deliv_df[deliv_df['batsman_runs']==6]
```

```python
#Number of sixes hit by each batsman
runs_6 = six.groupby('batsman')['batsman_runs'].count().reset_index()
runs_6.columns = ['batsman', 'sixes']
runs_6
```

```python
#Merging dataframes using concat
player = pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1],runs_6.iloc[:,1]], axis=1)
player
```

```python
#Filling null values with 0
player.fillna(0)
```

```python
#Adding strike_rate column
player['strike_rate'] = (player['batsman_runs'] / player['balls'])*100
```

```python
#Filtering dismissal kind for value 'caught'
deliv_df[deliv_df['dismissal_kind']=='caught']
```

```python
#Creating a list of dismissal kind for which bowler earns a wicket
dismissal_list = ['caught', 'bowled', 'lbw', 'caught and bowled', 'stumped', 'hit wicket']
```

```python
#Inserting matching values from deliv_df to dismissal_df
dismissal_df = pd.DataFrame()
for value in dismissal_list:
    temp_df = deliv_df[deliv_df['dismissal_kind']== value]
    dismissal_df = dismissal_df.append(temp_df)
```

```python
dismissal_df
```

```python
#Plot of bowlers with most wickets
most_wickets = dismissal_df.groupby('bowler')['bowler'].count()
most_wickets.sort_values(ascending=False)[:10].plot.bar()
```
